```python
from PIL import Image, ExifTags
import os

# Function to get image resolution
def get_image_resolution(image_path):
    with Image.open(image_path) as img:
        width, height = img.size
        return width, height

# Set the target resolution
target_resolution = (3000, 4000)

# Path to the folder containing images
folder_path = r'Images/'

# List all files in the directory
files = os.listdir(folder_path)

# Filter for image files (you can adjust the extensions as needed)
image_files = [f for f in files if f.lower().endswith(('jpg', 'jpeg', 'png', 'gif', 'bmp'))]

for file in image_files:
    image_path = os.path.join(folder_path, file)
    try:
        img = Image.open(image_path)
        
        # Rotate the image based on EXIF orientation
        for orientation in ExifTags.TAGS.keys():
            if ExifTags.TAGS[orientation] == 'Orientation':
                break
        exif = img._getexif()  # Use directly without calling items()

        if exif is not None:
            exif_orientation = exif.get(orientation)
            if exif_orientation == 3:
                img = img.rotate(180, expand=True)
            elif exif_orientation == 6:
                img = img.rotate(270, expand=True)
            elif exif_orientation == 8:
                img = img.rotate(90, expand=True)

        img = img.resize(target_resolution, resample=Image.LANCZOS)
        
        # Save to a different location
        output_path = os.path.join("Processed_Images", file)  # Save to a different folder
        img.save(output_path)

        # Check and print the resolution after processing
        processed_resolution = get_image_resolution(output_path)
        print(f"Processed Image - {file} Resolution: {processed_resolution}")
    except Exception as e:
        print(f"Failed to process {file}: {e}")
```
